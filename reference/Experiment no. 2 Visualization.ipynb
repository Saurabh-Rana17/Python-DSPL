```python
#Import dependencies 
import seaborn as sns
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
df = pd.read_csv("../self/clean4_car_price.csv") 
df['car_prices_in_rupee'] = pd.cut(df['car_prices_in_rupee'], bins=100, labels=False)

df2 = df
print(df2["car_prices_in_rupee"].unique()) 
```

```python
# Checking outliers
print(plt.figure(figsize = (10, 4), dpi = 100)) 
sns.boxplot(x = "car_prices_in_rupee", data = df2) 
```

```python
#Visualizing by stream

plt.figure(figsize = (10, 4), dpi = 100) 
color_palette = sns.color_palette("Accent_r") 
sns.set_palette(color_palette)
sns.countplot(x = "car_prices_in_rupee", data = df2) 
```

```python
#Count of internships done by students 
plt.figure(figsize = (10, 4), dpi = 100) 
color_palette = sns.color_palette("cool")
 
sns.set_palette(color_palette) 
sns.countplot(x = "car_prices_in_rupee", data = df2) 
```

```python
# # #Distribution of CGPA plt.figure(figsize = (10, 4), dpi = 100) 
grp = dict(df2.groupby('region').groups) 
m = {}
for key, val in grp.items(): 
    if key in m:
        m[key] += len(val) 
    else:
        m[key] = len(val) 
        plt.title("Distribution of Region") 
        plt.pie(m.values(), labels = m.keys())
```

```python
# #Exploring number of people by gender (Male/Female) 
plt.figure(figsize = (10, 4), dpi = 100)
color_palette = sns.color_palette("Accent_r") 
sns.set_palette(color_palette)
sns.countplot(x = "sex", data = df2) 
```

```python
# Region analysis of different bmi
plt.figure(figsize = (10, 4), dpi = 100) 
color_palette = sns.color_palette("magma") 
sns.set_palette(color_palette)
sns.barplot(x = "region", y = "bmi", data = df2) 
```

```python
# Region analysis as per charges 
plt.figure(figsize = (10, 4), dpi = 100) 
color_palette = sns.color_palette("plasma") 
sns.set_palette(color_palette)
 
sns.barplot(x = "region", y = "charges", data = df2) 
```

```python
# Smoking as per gender
df2["smoker"].replace({"yes":1, "no":0}, inplace = True) 
plt.figure(figsize = (10, 4), dpi = 100)
color_palette = sns.color_palette("plasma") 
sns.set_palette(color_palette)
sns.barplot(x = "sex", y = "smoker", data = df2) 
```

```python
# Number of smokers
df2["smoker"].replace({1:"yes", 0:"no"}, inplace = True) 
plt.figure(figsize = (10, 4), dpi = 100)
color_palette = sns.color_palette("BuGn_r") 
sns.set_palette(color_palette) 
sns.countplot(x = "smoker", data = df2)
```

```python
# Finding correlation 
plt.figure(figsize = (10, 6), dpi = 100) 
color = sns.color_palette("magma")
sns.heatmap(df2.corr(), vmax=0.9, annot=True,cmap = color)
```

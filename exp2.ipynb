```python
#Import dependencies 
import seaborn as sns
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
df = pd.read_csv("clean_car_price.csv") 
df2 = df
```

```python
#Checking outliers
print(plt.figure(figsize = (10, 4), dpi = 100)) 
sns.boxplot(x = "manufacture", data = df2) 
```

```python
#Visualizing by stream
plt.figure(figsize = (10, 4), dpi = 100) 
color_palette = sns.color_palette("Accent_r") 
sns.set_palette(color_palette)
sns.countplot(x = "car_prices_in_rupee", data = df2) 
```

```python
#Count of internships done by students 
plt.figure(figsize = (10, 4), dpi = 100) 
color_palette = sns.color_palette("cool")
 
sns.set_palette(color_palette) 
sns.countplot(x = "ownership", data = df2) 
```

```python
# # #Distribution of PRICES OF AIRLINE
plt.figure(figsize=(10, 4), dpi=100)
grp = df2.groupby('car_prices_in_rupee').size()
# Sorting and getting top N categories
N = 5 # Number of categories to display
top_N = grp.sort_values(ascending=False).head(N)
explode = (0,0,0,0,0) # Explode the 1st slice (if required)
# Plotting pie chart
plt.title("Prices of airline")
plt.pie(top_N, labels=top_N.index, explode=explode, autopct='%1.1f%%',startangle=140)
# Equal aspect ratio ensures that pie is drawn as a circle
plt.axis('equal')
plt.show()
```

```python
# #Exploring number of people by gender (Male/Female) 
plt.figure(figsize = (10, 4), dpi = 100)
color_palette = sns.color_palette("Accent_r") 
sns.set_palette(color_palette)
sns.countplot(x = "Seats", data = df2) 
```

```python
# Region analysis of different bmi
plt.figure(figsize = (10, 4), dpi = 100) 
color_palette = sns.color_palette("magma") 
sns.set_palette(color_palette)
sns.barplot(x = "Seats", y = "ownership", data = df2) 
```

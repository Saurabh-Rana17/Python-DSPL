```python
#Import dependencies 
import seaborn as sns
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
df = pd.read_csv("datasetForExp2AndExp3.csv")
df2 = df
```

```python
#Checking outliers
print(plt.figure(figsize = (10, 4), dpi = 100))
sns.boxplot(x="Voltage", data = df2)
```

```python
plt.figure(figsize=(10, 4), dpi=100)
sns.histplot(data=df2, x="Voltage", bins=30, kde=True, color='blue')
plt.xlabel("Voltage")
plt.ylabel("Count")
plt.title("Distribution of Voltage")
plt.show()
```

```python
# Did this because it was here, required was categorical data
plt.figure(figsize=(10, 4), dpi=100)
color_palette = sns.color_palette("cool")
sns.set_palette(color_palette)
sns.countplot(x="Voltage", data=df2)
plt.xlabel("Voltage")
plt.ylabel("Count")
plt.title("Distribution of Voltage")
plt.show()
```

```python
# Did this because it was here, required was categorical data
grp = dict(df2.groupby('Voltage').groups) 
m = {}
for key, val in grp.items(): 
    if key in m:
        m[key] += len(val) 
    else:
        m[key] = len(val) 
plt.figure(figsize=(8, 8))
plt.title("Distribution of Voltage")
plt.pie(m.values(), labels=m.keys(), autopct='%1.1f%%', startangle=140)
plt.show()
```

```python
# Did this because it was here, required was categorical data and numeric data
plt.figure(figsize=(10, 4), dpi=100)
color_palette = sns.color_palette("magma")
sns.set_palette(color_palette)
sns.barplot(x="Global_active_power", y="Voltage", data=df2)
plt.xlabel("Global Active Power")
plt.ylabel("Voltage")
plt.title("Relationship between Global Active Power and Voltage")
plt.show()
```

```python
non_numeric_columns = df2.select_dtypes(exclude=['float64', 'int64']).columns

# Exclude non-numeric columns from correlation analysis
numeric_df = df2.drop(columns=non_numeric_columns)

# Compute correlation matrix
correlation_matrix = numeric_df.corr()

# Plot heatmap
plt.figure(figsize=(10, 6), dpi=100)
color = sns.color_palette("magma")
sns.heatmap(correlation_matrix, annot=True, cmap=color, vmax=0.9)
plt.title("Correlation Heatmap")
plt.show()
```

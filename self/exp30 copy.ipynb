```python
#Import dependencies 
import seaborn as sns
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
df = pd.read_csv("clean4_car_price.csv") 
df2 = df
```

```python
df2.head()
```

```python
print(plt.figure(figsize = (10, 4), dpi = 100)) 
sns.boxplot(x = "car_prices_in_rupee", data = df2) 
```

```python
print(plt.figure(figsize = (10, 4), dpi = 100)) 
sns.boxplot(x = "car_prices_in_rupee", data = df2)
```

```python
percentile25 = df2['car_prices_in_rupee'].quantile(0.25)
percentile75 = df2['car_prices_in_rupee'].quantile(0.75)
iqr = percentile75 - percentile25
```

```python
upper_limit = percentile75 + 1.5 * iqr
lower_limit = percentile25 - 1.5 * iqr
print(upper_limit,lower_limit)
```

```python
df2[df2['car_prices_in_rupee'] > upper_limit]
# df2[df2['car_prices_in_rupee'] > lower_limit]
```

```python
new_df = df2[df2['car_prices_in_rupee'] > lower_limit]
new_df.shape
```

```python
plt.figure(figsize=(16,8))
plt.subplot(2,2,1)
sns.distplot(df2['car_prices_in_rupee'])
plt.subplot(2,2,2)
sns.boxplot(df2['car_prices_in_rupee'])
plt.subplot(2,2,3)
sns.distplot(new_df['car_prices_in_rupee'])
plt.subplot(2,2,4)
sns.boxplot(new_df['car_prices_in_rupee'])
plt.show()
```
